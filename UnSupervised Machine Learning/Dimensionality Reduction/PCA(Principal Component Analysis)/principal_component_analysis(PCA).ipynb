```python
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X,y = make_blobs(
    n_samples=1_000,
    n_features=4,
    centers = 4,
    random_state=42
)
```

### 4D Plot of 4 feature

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111,projection="3d")
ax.scatter(X[:,0],X[:,1],X[:,2],c=X[:,3])
plt.show()
```

```python
# Comparison between 3D plot & 4D plot
```

```python
fig = plt.figure(figsize=(10,6))

ax1 = fig.add_subplot(121,projection="3d")
ax1.scatter(X[:,0],X[:,1],X[:,2])
ax1.set_title("3D Scatter")

ax2 = fig.add_subplot(122,projection="3d")
sc = ax2.scatter(X[:,0],X[:,1],X[:,2],c=X[:,3])
ax2.set_title("3D Scatter (4th dim as color)")

plt.tight_layout()
plt.show()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2,random_state=42)

x_pca = pca.fit_transform(X)
```

```python
pca.explained_variance_ratio_
```

```python
pca.components_
```

```python
# Plot PCA
# Before PCA
plt.figure(figsize=(10,8))
plt.subplot(221)
plt.scatter(X[:,0],X[:,1],label=f"Shape :{X.shape}")
plt.legend()
plt.title("Before PCA")

# After PCA
plt.subplot(222)
plt.scatter(x_pca[:,0],x_pca[:,1],label=f"Shape:{x_pca.shape}")
plt.legend()
plt.title("After PCA")
plt.tight_layout()
plt.show()
```
