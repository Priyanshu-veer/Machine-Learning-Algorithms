```python
from sklearn.datasets import make_blobs, make_moons
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x,Y = make_moons(
    n_samples = 1_000,
    noise=0.05,
    random_state=42
)

sns.scatterplot(x=x[:,0], y = x[:,1])
plt.title("Moons Dataset")
plt.show()
```

*Try **DBSCAN** & **K-Means Clustering** on built-in dataset `make_moons.`*

```python
# use both k-means clustering & DSCAN
plt.figure(figsize=(15,10))
plt.subplot(221)
# K-Means Clustering
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2).fit_predict(x)

sns.scatterplot(x=x[:,0],y=x[:,1],c=kmeans)
plt.title("K-Means Clustering on make_moons Dataset")

# DBSCAN
from sklearn.cluster import DBSCAN
plt.subplot(222)
db = DBSCAN(eps=0.3,min_samples=5).fit_predict(x)

# plot
sns.scatterplot(x=x[:,0],y=x[:,1],c=db)
plt.title("DBSCAN on make_moons dataset")
plt.show()
```
