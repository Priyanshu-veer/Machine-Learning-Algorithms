```python
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
```

```python
X,y = make_blobs(
    n_samples=1_000,
    centers=4,
    n_features=2,
    cluster_std=1.3,
    random_state=42
)
```

```python
plt.scatter(x=X[:,0],y=X[:,1])
plt.xlabel("feature 1")
plt.ylabel("feature 2")
plt.show()
```

```python
agglo = AgglomerativeClustering(n_clusters=4).fit_predict(X)

plt.scatter(x=X[:,0],y=X[:,1],c=agglo)
plt.title("Agglomerativ Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram
```

```python
# linkage Matrix
L = linkage(X, method='ward')

plt.figure(figsize=(12, 6))
dendrogram(L)
plt.show()
```
