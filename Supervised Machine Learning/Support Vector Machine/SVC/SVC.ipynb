```python
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, f1_score, classification_report, confusion_matrix
```

```python
df = datasets.load_iris(as_frame=True).frame
```

```python
df.isnull().sum()
df.head()
```

```python
df['target'].value_counts()
```

```python
X = df.drop('target', axis = 1)
y = df['target']
# train & split
X_train, X_test, y_train, y_test = train_test_split(X,
                                                   y,
                                                   test_size = 0.3,
                                                   random_state = 42,
                                                   stratify=y)

# scale the data
ss = StandardScaler()
X_train_scaled = ss.fit_transform(X_train)
X_test_scaled = ss.transform(X_test)

# Support Vector Classifier
svc = SVC()
svc.fit(X_train, y_train)
# predicted values
y_pred = svc.predict(X_test)
# check accuracy of model
print(f'Accuracy Score : {accuracy_score(y_test, y_pred)}')
print(f'Classification Report : \n{classification_report(y_test, y_pred)}')
print(f'Confusion Matrix : \n{confusion_matrix(y_test, y_pred)}')
```

```python
# Check Overfitting & Underfitting

X_training_pred = svc.predict(X_train)
X_test_pred = svc.predict(X_test)

print(f'Training Accuracy Score : {accuracy_score(y_train,X_training_pred)}')
print(f'Testing Accuracy Score : {accuracy_score(y_test, X_test_pred)}')
```

```python
print(f"f1 score : {f1_score(y_train, X_training_pred, average='macro')}")
print(f"f1 score : {f1_score (y_test, X_test_pred,average='macro')}")
```

```python
print(f"f1 score : {f1_score(y_train, X_training_pred, average='weighted')}")
print(f"f1 score : {f1_score (y_test, X_test_pred,average='weighted')}")
```
