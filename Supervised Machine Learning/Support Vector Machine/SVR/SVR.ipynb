```python
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
from sklearn.metrics import r2_score
```

```python
df = datasets.load_diabetes(as_frame=True).frame
```

```python
df.head()
```

```python
X = df.drop('target', axis=1)
y = df['target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,
                                                   y,
                                                   test_size = 0.3,
                                                   random_state = 24)
```

```python
ss = StandardScaler()
y_train_scaled = ss.fit_transform(y_train.values.reshape(-1,1)).ravel()
y_test_scaled = ss.transform(y_test.values.reshape(-1,1)).ravel()
svr = SVR()

svr.fit(X_train, y_train_scaled)
y_pred = svr.predict(X_test)

print(f'r2 score : {r2_score(y_test_scaled, y_pred)}')
```

```python
# HyperParamete Tunning

from sklearn.model_selection import GridSearchCV

grid_params = {
    "C":[2,4,6,8,9,10],
    "kernel" : ["rbf","linear","poly"],
    "epsilon":[0.1,0.2,0.3,0.01,0.02,0.03]
}


svr = SVR()
grid_search = GridSearchCV(svr, grid_params, scoring="r2", cv=5)

grid_search.fit(X_train, y_train_scaled)
```
